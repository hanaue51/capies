```python
import numpy as np
from matplotlib import pyplot as plt
import math
```

```python
# データを折れ線グラフに表示する。

# 横軸: 年
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
# 縦軸: GDP
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]

# 第1引数: 横軸のデータ, 第2引数: 縦軸のデータ
# color: 線の色
# marker: 点の形
# linestyle: 線の書式
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')

# グラフの題名
plt.title("Nominal GDP")

# 縦軸のラベル
plt.ylabel("Billions of $")

# グラフを表示する。
plt.show()
```

```python
# 関数の値をグラフに表示する。

# 横軸の値を用意する。
# リストとして作成しても、NumPyの配列(ndarray)として作成してもよい。
# 下の例ではNumPyを使い、-piからpiまでの範囲で、pi/10間隔で値をとっている。
# piはNumPyで定義されている定数で、円周率(π)を値とする。
xs = np.arange(-np.pi, np.pi, np.pi / 10)
print(xs)

# 横軸の値それぞれに対して、縦軸の値を求めてグラフを描く。
# y = sin(x)
plt.plot(xs, np.sin(xs), label="y = sin(x)")
# y = cos(x)
plt.plot(xs, np.cos(xs), label="y = cos(x)")

# グラフの題名
plt.title("Trigonometric functions")
# 凡例
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

# ラベル
plt.xlabel("x")
plt.ylabel("y")

# グラフを表示する。
plt.show()
```

```python
# 自作の関数の値をグラフに表示する。

# ステップ関数
def step(x):
    return 1 if x > 0 else 0

# シグモイド関数
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# ランプ関数
def ramp(x):
    return x if x > 0 else 0

# 自作の関数をNumPyの配列へ適用できるようにする。
step_func = np.frompyfunc(step, 1, 1)
sigmoid_func = np.frompyfunc(sigmoid, 1, 1)
ramp_func = np.frompyfunc(ramp, 1, 1)

# 横軸の値を用意する。
xs = np.arange(-5, 5, 0.1)

# 横軸の値それぞれに対して、縦軸の値を求めてグラフを描く。
plt.plot(xs, step_func(xs), label="step")
plt.plot(xs, sigmoid_func(xs), label="sigmoid")
plt.plot(xs, ramp_func(xs), label="ramp")

# グラフの題名
plt.title("Threshold functions")
# 凡例
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)

# ラベル
plt.xlabel("x")
plt.ylabel("y")

# グラフを表示する。
plt.show()
```
